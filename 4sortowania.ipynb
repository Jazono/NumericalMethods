```python
import time

def sortbabel(n):
    tab1=tab
    start=time.time()
    
    for i in range(n-1):
        for i in range(n-1):
            if tab1[i]>tab1[i+1]:
                a=tab1[i]
                tab1[i]=tab1[i+1]
                tab1[i+1]=a
                
    finish=time.time()
    t1=finish-start
    return tab1,t1

def insertsort(n):
    tab2=tab
    start=time.time()
    
    for i in range(n):
        k=tab2[i]
        j=i-1
        while j>=0 and tab2[j]>k:
            tab2[j+1]=tab2[j]
            j=j-1
            tab2[j+1]=k
            
    finish=time.time()
    t2=finish-start
    return tab2,t2

def selectsort(n):
    tab3=tab
    start=time.time()
    
    i=0
    while i<n-1:
        minE=i
        for j in range(i+1,n):
            if tab3[j]<tab3[minE]:
                minE=j
        a=tab3[minE]
        tab3[minE]=tab3[i]
        tab3[i]=a
        i+=1
    
    finish=time.time()
    t3=finish-start
    return tab3,t3

def sortquick(tab):
    left = []
    center = []
    right = []

    if len(tab) > 1:
        pivot = tab[0]

        for n in tab:
            if n < pivot:
                left.append(n)
            elif n > pivot:
                right.append(n)
            else:
                center.append(n)
        return sortquick(left) + center + sortquick(right)
    else:
        return tab

def comparison(t1,t2,t3,t4):
    return(min(t1,t2,t3,t4))

print("Size")
n=int(input())

tab=[]
for i in range(n):
    tab.append(int(input()))
print(tab)

b=sortbabel(n)
print("bubblesort",b)

c=insertsort(n)
print("insertsort",c)

d=selectsort(n)
print("selectsort",d)

start = time.time()
result_sort_quick = sortquick(tab)
t4=time.time() - start
e = ("quicksort",result_sort_quick,t4)
print(e)


f=comparison(b[1],c[1],d[1],e[2])
print("shortest_time",f)
```
